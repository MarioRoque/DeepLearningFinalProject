# 3) Neural Network training.

### During this section of the notebook, the Neural Network will be feed the images using the next predefined hyperparameters:


- A 4 layer Model =  [12288, 20, 7, 5, 1]
- A learning rate = 0.0075
- A number of iterations = 2500
- Hidden layers activation function RELU.


This parameters were obtained from the original homework.

```python
### CONSTANTS ###
layers_dims = [12288, 20, 7, 5, 1] #  4-layer model
```

```python
# GRADED FUNCTION: L_layer_model

def L_layer_model(X, Y, layers_dims, learning_rate = 0.0075, num_iterations = 3000, print_cost=False, hidden_act_func="relu"):#lr was 0.009
    """
    Implements a L-layer neural network: [LINEAR->RELU]*(L-1)->LINEAR->SIGMOID.
    
    Arguments:
    X -- data, numpy array of shape (number of examples, num_px * num_px * 3)
    Y -- true "label" vector (containing 0 if cat, 1 if non-cat), of shape (1, number of examples)
    layers_dims -- list containing the input size and each layer size, of length (number of layers + 1).
    learning_rate -- learning rate of the gradient descent update rule
    num_iterations -- number of iterations of the optimization loop
    print_cost -- if True, it prints the cost every 100 steps
    
    Returns:
    parameters -- parameters learnt by the model. They can then be used to predict.
    """

    np.random.seed(1)
    costs = []                         # keep track of cost
    
    # Parameters initialization. 
    ### START CODE HERE ###
    parameters = initialize_parameters_deep(layers_dims)
    ### END CODE HERE ###
    
    # Loop (gradient descent)
    for i in range(0, num_iterations):

        # Forward propagation: [LINEAR -> RELU]*(L-1) -> LINEAR -> SIGMOID.
        ### START CODE HERE ### 
        AL, caches = L_model_forward(X, parameters, hidden_act_func)
        ### END CODE HERE ###
        
        # Compute cost.
        ### START CODE HERE ### 
        cost = compute_cost(AL, Y)
        ### END CODE HERE ###
    
        # Backward propagation.
        ### START CODE HERE ### 
        grads = L_model_backward(AL, Y, caches, hidden_act_func)
        ### END CODE HERE ###
 
        # Update parameters.
        ### START CODE HERE ### 
        parameters = update_parameters(parameters, grads, learning_rate)
        ### END CODE HERE ###
                
        # Print the cost every 100 training example
        if print_cost and i % 100 == 0:
            print ("Cost after iteration %i: %f" %(i, cost))
        if print_cost and i % 100 == 0:
            costs.append(cost)
            
    if(print_cost):
        # plot the cost
        plt.plot(np.squeeze(costs))
        plt.ylabel('cost')
        plt.xlabel('iterations (per tens)')
        plt.title("Learning rate =" + str(learning_rate))
        plt.show()
    
    return parameters
```

# 5) Parameters exploration.

During this section, a hyperparameter exploration will be performed.

This considering a list of values for each of the layers on the Neural network (This list will check for all the configurations for layers between 3 to 7 (Included input and output layers predefined), a range of iterations, and a range of learning rates.

Run the next two cells to obtain the configurations for all the layers.

```python
# Reasignin plot sizes
%matplotlib inline
plt.rcParams['figure.figsize'] = (5.0, 4.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'
```

```python
decode = [10, 15, 20]

def rec(conv_list, min_V, max_V, max_L, L):
    if(L >= max_L):
        global_conv_list.append(conv_list)
        return
        
    for val in range(min_V, max_V +1):
        rec( [decode[val]] + conv_list, min_V, max_V, max_L, L+1)

global_conv_list = []
for layer_num in range(1,5):
    rec([],0, 2, layer_num, 0)
    
for row in range(len(global_conv_list)):
    global_conv_list[row] = [12288, 10] + global_conv_list[row] + [1]

combinations = len(global_conv_list)
print("Total combinations of layers and neurons: ", combinations)
```

As it can be seen, the total convinations for layers and neurons on each layer are 120. This convination will be trained on by variating the other defined hyper parameters to obtain the one that suits the best the Mask vs No-Mask problem.

On this section you will: Define a function to calculate the acurracy, train the Neural Network on 120 combinations of layers, from 1000 to 1500 iterations, and a learning rate that goes from 1/10 to 51/10. 

This will write into a csv file in order to avoid the user to train the N-N each time the program runs, thus allowing to analyze data latter.

```python
def calculateAcurracy(p,y):
    m = p.shape[1]
    return str(np.sum((p == y)/m))
```

# NOTE: Do not run this line if you do not want to wait for the 12 hours to finish running, simply skip this code cell. And go to the 6)

```python
results = []
step =0
learning_rate = 50
iterations = 100

step_learning =10
lower_learning = 1
upper_learning = 51

step_iterations = 100
lower_iterations = 1000
upper_iterations = 1500


total_steps =  (upper_learning-lower_learning)//step_learning*(upper_iterations-lower_iterations)//step_iterations*combinations

for learning_rate in range(lower_learning, upper_learning, step_learning):
    for iterations in range(lower_iterations,upper_iterations,step_iterations):
        for config in global_conv_list:
            layers_dims = config
            parameters = L_layer_model(train_x, train_y, layers_dims=config,learning_rate = learning_rate/1000, num_iterations = iterations, print_cost = False)
            pred_dev = predict(dev_x, dev_y, parameters)
            acc = calculateAcurracy(dev_y, pred_dev)

            results.append([config, learning_rate/1000, iterations, acc])
            step += 1
            print("We are in step:", step, "of ",total_steps, "steps (", (step/total_steps*100), "% )\n")
```

```python
# Writes the ordered results to a csv file
df_result = pd.DataFrame(results, columns=["Configs","learning_rate","iterations","acc"])
df_result = df_result.sort_values(by=["acc"], ascending=False)
df_result.to_csv("Parameter_exploration.csv", index=False)
```

As it can be seen, the index 0,5, 7,9, and 11 are the best architectures, because they get a great acurracy and they require the less iterations and the smallest number of layers.

Using this, a new set of neural networks will be trained, by variating ht eactivating layer between TANH and RELU.

Run the next cell to being exploration:

```python
# The values of the list are the indexes of the best 5 configurations from the df_result
top_5_architectures_index = [0,5,7,9,11]
index_and_accuracy = []

for index in top_5_architectures_index:
    print("Index Model: ", index)
    print("\nTANH:")
    parameters = L_layer_model(train_x, train_y, 
                               layers_dims = df_result.iloc[index].Configs,
                               learning_rate = df_result.iloc[index].learning_rate, 
                               num_iterations = df_result.iloc[index].iterations, 
                               print_cost = True,
                               hidden_act_func="tanh")
    
    pred_dev_1 = predict(dev_x, dev_y, parameters)
    acc_tanh = calculateAcurracy(dev_y, pred_dev_1)
    
    
    print("\nRELU:")
    parameters = L_layer_model(train_x, train_y, 
                               layers_dims = df_result.iloc[index].Configs,
                               learning_rate = df_result.iloc[index].learning_rate, 
                               num_iterations = df_result.iloc[index].iterations, 
                               print_cost = True,
                               hidden_act_func="relu")
    
    pred_dev_2 = predict(dev_x, dev_y, parameters)
    acc_relu = calculateAcurracy(dev_y, pred_dev_2)
    
    index_and_accuracy.append([index, acc_tanh, "tanh"])
    index_and_accuracy.append([index, acc_relu, "relu"])
```
